```python
from pysentimiento import create_analyzer #Importa a função "create_analyzer" da biblioteca "pysentimiento" instalada anteriormente. Ela é respondsavel por criar o modelo de analize de sentimento.
modelo_analise_sentimento = create_analyzer(task='sentiment', lang='pt') #Cria um modelo de analise de sentimento usando a função "create_analyzer". O Parametro "task" especifica que devera ser realizado uma analise de sentimento, e o parametro "lang", especifica qual linguagem será realizada a analize (Português).
```

```python
import pandas as pd #Importa a biblioteca pandas com o alias/apelido de "pd"
df_resenhas_produto = pd.read_csv('resenhas_produto.csv') # importa o 'resenhas_produto.csv' e o lê como um DataFrame do pandas
df_resenhas_produto # mostra o conteudo do dataframe criado
```

```python
Predict_Output = modelo_analise_sentimento.predict(df_resenhas_produto['Resenha']) # usa 'modelo_analise_sentimento' para prever o sentimento de cada resenha no 'df_resenhas_produto'
Predict_Output # mostra os resultados das previsoes
```

```python
sentimento = [] # Inicializa um array vazio para armazenar os sentimentos
for resultado in Predict_Output: # percorre os resultados das previsões do modelo
  sentimento.append(resultado.output) # Adiciona o atributo 'output' (saida de cada resenha, positivo, negativo ouu neutro) de cada resultado na lista 'sentimento'
sentimento # mostra
```

```python
df_resenhas_produto['Sentimento'] = sentimento # cria a coluna 'Sentimento' no DataFrame e atribui a lista 'sentimento' a essa coluna
df_resenhas_produto
```

```python
df_sentimento = df_resenhas_produto.groupby('Sentimento').value_counts().reset_index(name='Contagem') # agrupa o DataFrame por 'Sentimento' faz um count em  de cada combinação de 'Sentimento' e resenha
df_sentimento.head() #mostra os 5 primeiros
```

```python
import plotly.express as px # importa a biblioteca Plotly Express (gerar de gráficos)
fig = px.bar(df_sentimento, x='Sentimento', y='Contagem', title='Contagem de Resenhas por Sentimento') #cria um gráfico de barras com df_sentimento, que mostra a contafem de resenha por sentimento
fig.show() #mostra o grafico criado
```

```python
import nltk # Importa a biblioteca NLTK (processamento de linguagem natural)
from nltk.corpus import stopwords # Importa a lista de stopwords presentes no corpus NLTK
```

```python
# Carregar stopwords em português
nltk.download('stopwords')
portuguese_stopwords = set(stopwords.words('portuguese'))
```

```python
from wordcloud import WordCloud # importa a classe WordCloud pra gerar a nuven de palavras
import matplotlib.pyplot as plt # importa o matplotlib (gráficos/imagens)
```

```python
def nuvem_palavras(texto, coluna_texto, sentimento):
# Filtrando as resenhas com base no sentimento especificado
  texto_sentimento = texto.query(f"Sentimento == '{sentimento}'")[coluna_texto]
  # Unindo todas as resenhas em uma única string
  texto_unido = " ".join(texto_sentimento)
  # Dividindo o texto em palavras e filtrando com stopwords
  palavras = texto_unido.split()
  palavras_filtradas = [palavra for palavra in palavras if palavra not in portuguese_stopwords]
  texto_filtrado = " ".join(palavras_filtradas)
  # Criando e exibindo a nuvem de palavras
  nuvem_palavras = WordCloud(width=800, height=500, max_words=50).generate(texto_filtrado)
  plt.figure(figsize=(10, 7))
  plt.imshow(nuvem_palavras, interpolation='bilinear')
  plt.axis('off')
  plt.show()
```

```python
nuvem_palavras(df_resenhas_produto, 'Resenha', 'POS') #Mostra nuvem de palavras frequentes em analises Positivas
```

```python
nuvem_palavras(df_resenhas_produto, 'Resenha', 'NEG') #Mostra nuvem de palavras frequentes em analises negativas
```

```python
nuvem_palavras(df_resenhas_produto, 'Resenha', 'NEU') #Mostra nuvem de palavras frequentes em analises NEUTRAS
```
