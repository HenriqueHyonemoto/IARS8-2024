```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import pandas as pd

# Carregar o modelo e o tokenizador
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

# Dados fictícios para o status dos pedidos
dados_pedidos = {
    "numero_pedido": ["12345", "67890", "11121", "22232"],
    "status": ["Shipped", "Processing", "Delivered", "Cancelled"]
}

df_status_pedidos = pd.DataFrame(dados_pedidos)

# Função para verificar o status do pedido
def verificar_status_pedido(numero_pedido):
    try:
        status = df_status_pedidos[df_status_pedidos["numero_pedido"] == numero_pedido]["status"].iloc[0]
        return f"O status do seu pedido {numero_pedido} é: {status}"
    except:
        return "Número do pedido não encontrado. Por favor, verifique e tente novamente."

# Função principal de resposta do chatbot
def respondendo_faq(input_usuario, chat_history_ids=None):
    palavras_chave_status = ["pedido", "status do pedido", "meu pedido", "verificar pedido", "atualização do pedido"]

    # Se o usuário pedir o status de um pedido
    if any(keyword in input_usuario.lower() for keyword in palavras_chave_status):
        # Pedir o número do pedido
        return "Por favor, forneça o número do seu pedido.", chat_history_ids

    # Caso o usuário tenha inserido um número de pedido
    if input_usuario.isdigit():
        return verificar_status_pedido(input_usuario), chat_history_ids

    # Caso contrário, o chatbot deve gerar uma resposta com o modelo DialoGPT
    novo_usuario_input_ids = tokenizer.encode(input_usuario + tokenizer.eos_token, return_tensors='pt')

    if chat_history_ids is not None:
        bot_input_ids = torch.cat([chat_history_ids, novo_usuario_input_ids], dim=-1)
    else:
        bot_input_ids = novo_usuario_input_ids

    chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    resposta = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)

    return resposta, chat_history_ids

import gradio as gr

# Criar a interface Gradio
app = gr.Interface(
    fn=respondendo_faq,
    inputs=[gr.Textbox(label="Digite sua mensagem", placeholder="Pergunte algo ou informe o número do seu pedido..."), gr.State()], # Added gr.State() to inputs
    outputs=[gr.Textbox(label="Resposta do Bot"), gr.State()],
    title="ChatBot com Status de Pedido",
    description="Este chatbot pode responder sobre status de pedidos ou bater papo sobre outros tópicos.",
)

# Iniciar a aplicação
if __name__ == "__main__":
    app.launch(share=True)
```
