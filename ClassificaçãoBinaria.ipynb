```python
# features (! sim, 0 nao)
# pelo longo.?
# perna curta?
# late?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]
cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

# 1 => porco, 0 => cachorro
dados =[porco1, porco2, porco3, cachorro1, cachorro2, cachorro3] #dados contem as caracteristicas dos animais
classe = [1,1,1,0,0,0]

# do modulo sklearn.vm, importaremos um estimador chamado LinearSVC
from sklearn.svm import LinearSVC

# criando o estimador usando LinearSVC
# A classe LinearSVC implementa um algoritimo de cassificação, o SVM, ou máquina de vetores de su
modelo = LinearSVC()
# treinar o modelo
modelo.fit(dados, classe) # fit = treinar, variavel X são todas as variaveis, variavel Y é a que define se ele é ou não (leitura das variaveis X)
# descobrir se o animal misterioso é porco ou cachorro
animal_misterioso = [1,1,1]
# classificar o animal misterioso
adivinhar = modelo.predict([animal_misterioso]) #predict faz previsão
#podemos fazer isto para diversos animais misteriosos para a predição
print (adivinhar)
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
testes = [misterio1,misterio2,misterio3]
previsoes = modelo.predict(testes)
print(previsoes)
# retorna a saida [0 1 0] -> cachorro, porco, cachorro
# mas se tratavam de um cachorro e dois porcos ([0,1,1]), neste caso, o açgoritimo errou 1 de 3.
# qual seria a resposta correta (0,1,0)
# criar a estrutura que representa a resposta correta (0,1,1) para comparar com a resposta da variavel
real = [0,1,1]
#calcular a taxa de acerto da biblioteca sklearn.metrics -> accuracy_score
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(real, previsoes)
print("Taxa de acerto: %.2f" % (taxa_de_acerto * 100))


```
